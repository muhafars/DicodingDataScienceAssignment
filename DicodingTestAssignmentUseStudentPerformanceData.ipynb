## Data Wrangling

```python
data = pd.read_csv('student_performance_data.csv', delimiter=';')
print(data.head())
```

## Exploratory Data Analysis (EDA)
### Explore

```python
fig= px.histogram(data, y = 'Status', nbins=20, title='College Status')
fig.show()
```

```python
fig = px.box(data, x = 'Unemployment_rate', title = 'Tingkat Pengangguran')
fig.show()
```

```python
fig = px.histogram(data,  x= 'GDP', title= 'Gross Domestic Products')
fig.show()
```

```python
fig = px.box(data, x='Inflation_rate', title='Inflation Rate Plot Box')
fig.show()
```

## Visualization & Explanatory Analysis
### Pertanyaan 1

```python
fig = px.histogram(data, x = 'Status', y = 'Unemployment_rate', title= 'Histogram Level Unemployment Rate of Status Graduate')
fig.show()
```

```python
fig = px.bar(data, y = 'Status', x = 'Unemployment_rate', title= 'Level Unemployment Rate of Status Graduate')
fig.show()
```

### Pertanyaan 2

```python
fig = px.scatter(data, y='GDP', x='Inflation_rate',  title='Relation GDP & Inflation Rate', trendline="ols")
fig.show()
```
